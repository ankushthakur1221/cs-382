```python
def mp_neuron(x1, x2, threshold):
    total_sum = x1 + x2

    if total_sum >= threshold:
        return 1
    else:
        return 0
```

```python
t_and = 2

print("AND Gate Results (Threshold = 2):")
print("0, 0 ->", mp_neuron(0, 0, t_and))
print("0, 1 ->", mp_neuron(0, 1, t_and))
print("1, 0 ->", mp_neuron(1, 0, t_and))
print("1, 1 ->", mp_neuron(1, 1, t_and))
```

```python
t_or = 1

print("OR Gate Results (Threshold = 1):")
print("0, 0 ->", mp_neuron(0, 0, t_or))
print("0, 1 ->", mp_neuron(0, 1, t_or))
print("1, 0 ->", mp_neuron(1, 0, t_or))
print("1, 1 ->", mp_neuron(1, 1, t_or))
```
